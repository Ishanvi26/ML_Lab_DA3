### Importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = iris.data
```

### Standardizing the dataset

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Fit K-means with 3 clusters but stop after 5 iterations (epochs)

```python
kmeans = KMeans(n_clusters=3, max_iter=1, random_state=42, init="random", n_init=1)
```

### Use PCA to reduce dimensions for visualization

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

### Plotting the clusters after fitting the model for every epoch

```python
plt.figure(figsize=(10, 7))
for epoch in range(1, 6):
    kmeans.fit(X_scaled)
    labels = kmeans.labels_  # Cluster assignments
    centroids = kmeans.cluster_centers_  # Centroids after current iteration
    centroids_pca = pca.transform(centroids)

    # Plot the data points and centroids
    plt.subplot(2, 3, epoch)
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap="viridis", s=50, label="Data points")
    plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c="red", s=200, marker="X", label="Centroids")
    plt.title(f"K-Means Clustering: Iteration {epoch}")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.legend()

plt.tight_layout()
plt.show()
```
